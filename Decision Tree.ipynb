```python
data= pd.get_dummies(df1,columns=['purpose'],drop_first=True)
data
```

```python
# split the data into dependent and independent 
x=data.drop('not.fully.paid',axis=1)
y=data['not.fully.paid']
```

```python
# split data into train and test set 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=100)
```

```python
#train the decision tree 
from sklearn.tree import DecisionTreeClassifier
dtree= DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```python
y_pred=dtree.predict(x_test)
y_pred
```

```python
y_test
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_test,y_pred))
```

```python
# accuracy score 
print(accuracy_score(y_test,y_pred))
```

# Hyperparameter tuning 
the goal of hyper parameter tuning is to find the values that lead to the best performance on a given task 
# what is hyperparameter?
in the context of machine learning ,hyperparameters are configuration variables that are set before the training process of a model begins. they control the learning process itself ,rather than being learned from the data . hyperparameteers are often used to tune the performance of a model ,and they can have a significant impact on the model accuracy ,generalization and other metrics .

# hyperparameter tuning techniques 
models can have hyperparameters and finding the best combination of parameters can be treated as a search problem .the best two strategies of hyperparameter tuning are :
gridsearchCV  RandomSearch

```python
# hyperparameter tuning gridsearch 
from sklearn.model_selection import GridSearchCV
param_grid={'max_depth':[2,3,4,5],
           'max_features':[1,3,5,10],
           'min_samples_split':[2,3,5,10],
           'criterion':['gini','entropy'],
           'min_samples_leaf':[1,2,3,4]
           }
grid=GridSearchCV(dtree,param_grid=param_grid,verbose=2)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
# train decision tree 
dtree =DecisionTreeClassifier(criterion= 'entropy',
 max_depth= 3,
 max_features= 1,
 min_samples_leaf= 2,
 min_samples_split= 5)
dtree.fit(x_train,y_train)
```

```python
y_pred=dtree.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
